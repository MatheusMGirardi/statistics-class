```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('seaborn-v0_8')
```

```python
df = pd.read_csv('dataset.csv', index_col='Rank')
```

```python
df = df.dropna()
```

```python
df.head()
```

```python
sales_per_year = df.groupby('Year')[['Global_Sales', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sum().reset_index()
sales_per_year
```

### 1. Gráfico em linha

```python
plt.figure(figsize=(10, 6))
plt.plot(sales_per_year['Year'], sales_per_year['Global_Sales'], linestyle='-', color='red', label='Global Sales')

plt.legend()
plt.title(f"Global sales of video games by year")
plt.xlabel("Year")
plt.ylabel("Global Sales (millions of units)")
plt.grid()
plt.show()
```

### Análise:
- Podemos ver que há um crescimento no número de jogos vendidos de acordo com os anos e depois de 2010 o dataset bem possivelmente não manteve todos os jogos que foram lançados.


## 2. Gráfico de Área

```python
plt.figure(figsize=(8, 5))
alpha=0.4
plt.fill_between(sales_per_year['Year'], sales_per_year['Global_Sales'], linestyle='-', color='red', label='Global Sales', alpha=alpha)
plt.fill_between(sales_per_year['Year'], sales_per_year['EU_Sales'], linestyle='-', color='blue', label='EU Sales', alpha=alpha)
plt.fill_between(sales_per_year['Year'], sales_per_year['NA_Sales'], linestyle='-', color='yellow', label='NA Sales', alpha=alpha)
plt.fill_between(sales_per_year['Year'], sales_per_year['JP_Sales'], linestyle='-', color='green', label='JP Sales', alpha=alpha)
plt.fill_between(sales_per_year['Year'], sales_per_year['Other_Sales'], linestyle='-', color='cyan', label='Other Sales', alpha=alpha)
plt.legend()
plt.title(f"Area graph")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.grid() 
plt.show()
```

## Análise
- Podemos perceber que há um padrão de crescimento que se mantém nas diferentes regiões exceto o Japão que parece se manter.

## 3. Gráfico de Barras

```python
    plt.figure(figsize=(10, 5))
    plt.bar(sales_per_year['Year'], sales_per_year['Global_Sales'])
    plt.title(f"Global sales of video games by year")
    plt.xlabel("Year")
    plt.ylabel("Global Sales (millions of units)")
    plt.grid()
    plt.show()
```

### Análise
- Podemos perceber melhor alguns pardrões nos dados, como o padrão sinoidal de 1980 à 1995. Assim como comparar melhor o volume de vendas entre os anos.

## 4. Histograma

```python
plt.figure(figsize=(10, 6))
plt.hist(df['JP_Sales'], bins=5, color='orange', edgecolor='black')
plt.title('Distribution of Japan Sales (Histogram)')
plt.xlabel('Japan Sales (millions)')
plt.ylabel('Frequency')
plt.show()
```

```python
### Análise
- No histograma observamos que, no Japão, pouquissímos jogos foram capazes de ultrapassar a marca de 2 milhões de cópias vendidas,
```

## 5. Gráfico de Dispersão

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Year'], df['Global_Sales'], s=df['Global_Sales']*10, c='red', alpha=0.6)
plt.title('Global Sales by Year (Scatter Chart)')
plt.xlabel('Release Year')
plt.ylabel('Global Sales (millions)')
plt.grid(True)
plt.show()
```

### Análise 
- Percebe-se um padrão semelhante ao do gráfico de linha mas evidenciando o quão incomum foram as vendas entre 2005 e 2010

## 6. Gráfico de Caixa

```python
plt.figure(figsize=(10, 6))
df.boxplot(column='NA_Sales', by='Genre', grid=False)
plt.title('Distribution of North America Sales by Genre (Box Chart)')
plt.suptitle('')  # Remove the default pandas title
plt.xlabel('Genre')
plt.ylabel('NA Sales (millions)')
plt.xticks(rotation=45, ha='right')
plt.show()
```

### Análise
- Nesse caso o boxplot não trouxe muitras informações interessantes dado a distribuição dos dados. Testamos com outras variavéis sem sucesso

## Gráfico de Torta

```python
plt.figure(figsize=(15, 15))
platform_sales = df.groupby('Platform')['Global_Sales'].sum()
plt.pie(platform_sales, labels=platform_sales.index, autopct='%1.1f%%', startangle=90)
plt.title('Global Sales Distribution by Platform (Pie Chart)')
plt.show()
```

### Análise
-  Aqui podemos ver o quanto cada plataforma representa de todas as vendas. Observando que certos consoles tiveram grande impacto (como PS2, Xbox 360 e Wii)
